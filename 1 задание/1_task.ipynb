```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
N = 100
tetta = 121
selection = [random.randint(0, 121) for i in range(N)]
```

```python
tetta_1 = [2*sum(selection[:n])/(n+1) for n in range(1, N+1)]
tetta_2 = [(n+1)*min(selection[:n]) for n in range(1, N+1)]
tetta_3 = [((n+1)*(max(selection[:n])))/n for n in range(1, N+1)]
tetta_4 = [min(selection[:n]) + max(selection[:n]) for n in range(1, N+1)]
tetta_5 = [selection[:n][0] + sum(selection[1:n])/(n-1) for n in range(2, N+1)]
tetta_5.insert(0, selection[0])
```

```python
y1 = [abs(elem - tetta) for elem in tetta_1]
y2 = [abs(elem - tetta) for elem in tetta_2]
y3 = [abs(elem - tetta) for elem in tetta_3]
y4 = [abs(elem - tetta) for elem in tetta_4]
y5 = [abs(elem - tetta) for elem in tetta_5]
```

```python
x = [n+1 for n in range(N)]
```

```python
fig, axes = plt.subplots(1, 2, figsize = (15, 5))
axes[0].plot(x, y1, color='r', label = '2*middle(x)')

axes[0].plot(x, y3, color='b', label = 'max')
axes[0].plot(x, y4, color='c', label = 'min+max')
axes[0].plot(x, y5, color='m', label = 'x1 + middle')
axes[0].legend()
axes[0].grid(True)
axes[0].minorticks_on()
axes[0].grid(which='minor')

axes[1].plot(x, y2, color='g', label = 'min')
axes[1].legend()
axes[1].grid(True)
axes[1].minorticks_on()
axes[1].grid(which='minor')

plt.show()
```
